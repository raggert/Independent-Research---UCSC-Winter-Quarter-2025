```python
def get_param_arrays(filename):
    data = np.load(filename +'.npz', allow_pickle=True)
    
    popt_mbb = data['popt_mbb']
    pcov_mbb = data['pcov_mbb']
    popt_myy = data['popt_myy']
    pcov_myy = data['pcov_myy']
    
    data.close()

    return popt_mbb, pcov_mbb, popt_myy, pcov_myy
```

```python
def get_N(popt, pcov):
    N = ufloat(popt[0], np.sqrt(pcov[0][0]))

    return N
```

```python
def get_mu(popt, pcov, i):
    mu = ufloat(popt[i][0], np.sqrt(pcov[i][0][0]))

    return mu
```

```python
def get_sigma(popt, pcov, i):
    sigma = ufloat(popt[i][1], np.sqrt(pcov[i][0][0]))
    return sigma
```

```python
popt_mbb_Y1, pcov_mbb_Y1, popt_myy_Y1, pcov_myy_Y1 = get_param_arrays('HM-SampleArrays/HM-Y=1/Y=1-Fitparams')

popt_mbb_Y2, pcov_mbb_Y2, popt_myy_Y2, pcov_myy_Y2 = get_param_arrays('HM-SampleArrays/HM-Y=2/Y=2-Fitparams')

popt_mbb_Y3, pcov_mbb_Y3, popt_myy_Y3, pcov_myy_Y3 = get_param_arrays('HM-SampleArrays/HM-Y=3/Y=3-Fitparams')
```

## Mean

### VBFHH

#### m$_{b\bar{b}}$

```python
VBFHH_mbb_mu_Y1 = get_mu(popt_mbb_Y1, pcov_mbb_Y1, 0)
VBFHH_mbb_mu_Y2 = get_mu(popt_mbb_Y2, pcov_mbb_Y2, 0)
VBFHH_mbb_mu_Y3 = get_mu(popt_mbb_Y3, pcov_mbb_Y3, 0)

VBFHH_mbb_mu_arr = np.array([VBFHH_mbb_mu_Y1, VBFHH_mbb_mu_Y2, VBFHH_mbb_mu_Y3])
```

```python
plt.title(r'$\mu_{peak}$ at different sensitivities (VBFHH - m$_{b\bar{b}}$)')

plt.xticks([1,2,3],[1,2,3])
plt.yticks(np.linspace(95e3,115e3,5),(np.linspace(95e3,115e3,5)/1e3).astype(float))

plt.xlabel('sensitivity')
plt.ylabel('$\mu_{peak}$ [GeV]')

plt.errorbar(np.arange(1,4), unumpy.nominal_values(VBFHH_mbb_mu_arr), yerr=unumpy.std_devs(VBFHH_mbb_mu_arr),\
            ls='', marker='x', capsize=2, color='blue')

#plt.savefig('./Plots/Fitparameter-Comparisons/VBFHH_mbb_mu.pdf', format='pdf')
```

## Normalization 2D

```python
def get_param_2d_arrays(filename):
    data = np.load(filename +'.npz', allow_pickle=True)
    
    popt = data['popt']
    pcov = data['pcov']
    
    data.close()

    return popt, pcov
```

```python
popt_signal_2d_Y1, pcov_signal_2d_Y1 = get_param_2d_arrays('HM-SampleArrays/HM-Y=1/Y=1-2D-Signalfit')

popt_signal_2d_Y2, pcov_signal_2d_Y2 = get_param_2d_arrays('HM-SampleArrays/HM-Y=2/Y=2-2D-Signalfit')

popt_signal_2d_Y3, pcov_signal_2d_Y3 = get_param_2d_arrays('HM-SampleArrays/HM-Y=3/Y=3-2D-Signalfit')
```

```python
signal_2d_N_Y1 = get_N(popt_signal_2d_Y1, pcov_signal_2d_Y1)
signal_2d_N_Y2 = get_N(popt_signal_2d_Y2, pcov_signal_2d_Y2)
signal_2d_N_Y3 = get_N(popt_signal_2d_Y3, pcov_signal_2d_Y3)

signal_2d_N_arr = np.array([signal_2d_N_Y1, signal_2d_N_Y2, signal_2d_N_Y3])
```

```python
plt.title('Normalization of Signal (2D-Fit)')
plt.xlabel('Sensitivity')
plt.ylabel(r'$\mu$')

plt.ylim(-25,25)

plt.xticks([1,2,3],[1,2,3])

plt.errorbar(np.arange(1,4), unumpy.nominal_values(signal_2d_N_arr), yerr=unumpy.std_devs(signal_2d_N_arr), ls='',\
            marker='x', capsize=2, color='red')


plt.grid(ls=':')
plt.axhline(1, ls='dashed', color='grey');
```
